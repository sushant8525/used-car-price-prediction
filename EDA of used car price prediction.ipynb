## Handelling of null values


```python
##fill NAN
df['Mileage']=df['Mileage'].fillna(method="ffill")
df['Engine']=df['Engine'].fillna(method="ffill")   
df['Power']=df['Power'].fillna(method="ffill")      
df['Seats']=df['Seats'].fillna(method="ffill")      
```

```python
#to check is any NAN value remain
df.isna().sum()
```

### drop some extra characters and alphabets in dataframe 

```python
df['Mileage'] = df['Mileage'].str[0:4]
```

```python
df['Engine'] = df['Engine'].str[0:4]
```

```python
df['Power'] = df['Power'].str[0:-3]
```

```python
#slicing as we required manufacture name
df1 = df["Name"].str.split(" ", expand = True)
```

```python
df["Manufacturer"] = df1[0]
```

```python
#decide which columns we want and which ones we drop from our analysis
df.drop(['New_Price','Unnamed: 0','Name'], axis=1,inplace=True)
```

### Year



```python
#Year has no significance on its own unless we try to extract how old a car is from this and see how its resale price may get affected.
curr_time = datetime.datetime.now()
df['Year'] = df['Year'].apply(lambda x : curr_time.year - x)
```

# Feature Engineerin

```python
df["Mileage"] = [float(str(i).replace(",", "")) for i in df["Mileage"]]
```

```python
df["Fuel_Type"] = df["Fuel_Type"].replace({'Diesel':1,'Petrol':2,'CNG':3,'LPG':4,'Electric':5})
```

```python
df["Transmission"] = df["Transmission"].replace({'Manual':1,'Automatic':2})
```

```python
df["Owner_Type"] = df["Owner_Type"].replace({'First':1,'Second':2,'Third':3,'Fourth & Above':4})
```

```python
df['Engine']=df['Engine'].astype('int64')
```

```python
df['Mileage']=df['Mileage'].astype('float')
```
